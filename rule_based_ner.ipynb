## Rules based NER
For this, we will simply check the text with list of characters to find the characters that occur in the text

```python
import json
import re
```

```python
with open("data/hp_text.txt","r", encoding="utf-8") as f:
    text = f.read().split("\n\n")

character_names = []
with open("data/hp_characters.json","r", encoding="utf-8") as f:
    characters = json.load(f)
    for character in characters:
        names  = character.split()
        for name in names:
            if "and" != name and "the" != name and "The" != name:
                name = name.replace(",", "").strip()
                character_names.append(name)
```

```python
# regex for removing punctuations
pattern = r'[^a-zA-Z0-9\s]'
for segment in text:
    segment =  segment.strip()
    # remove punctuation
    segment = re.sub(pattern, '', segment)
    words = segment.split()
    print(words)

    i = 0
    for word in words:
        if word in character_names:
             print (f"Found Character(s): {word}")

        i=i+1
```
